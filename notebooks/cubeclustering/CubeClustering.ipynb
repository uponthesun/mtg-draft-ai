```python
lines = []

for i in range(1, 11):
    with open('cubes{}.txt'.format(i), 'r') as f:
        lines += f.readlines()
```

```python
import json
from json import JSONDecodeError

all_cards = set()

cubes = [json.loads(l) for l in lines]

for c in cubes:
    all_cards.update(c)
```

```python
len(all_cards)
```

```python
all_cards_sorted = sorted(list(all_cards))
```

```python
index_by_name = {name: index for (index, name) in enumerate(all_cards_sorted)}
```

```python
import itertools

shared_cube_count = {}

for cube in cubes:
    for pair in itertools.combinations(cube, 2):
        shared_cube_count[pair] = shared_cube_count.setdefault(pair, 0) + 1
```

```python
shared_cube_count
```

```python
sorted_pairs = sorted(list(shared_cube_count.items()), key=lambda pair: pair[1], reverse=True)
```

```python
len(sorted_pairs)
```

```python
ancestral_pairs = [pair for pair in sorted_pairs if pair[0][0] == 'Ancestral Recall']
```
